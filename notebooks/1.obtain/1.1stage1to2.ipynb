### Introduction:

This is a simple notebook that takes our .csvs and combines them into one large .csv file. This script acts
as a dataframe collector, concatenating all of the separate dataframes in stage 1.

```python
import json
import glob
import pandas as pd
import os
```

```python
tweetPath = "../../data/1.stage1/tweets/"
userPath = "../../data/1.stage1/users/"

enumTweets = glob.glob(tweetPath+"*")
enumUsers = glob.glob(userPath+"*")
```

```python
enumTweets
enumUsers
```

```python
### Support Code:

def concatfile(listPaths,subdir):
    fileMap={}
    for path in listPaths:
        index = int(path.replace(subdir,"").replace(".csv",""))
        fileMap[index] = path
    
    return fileMap

def mergefiles(pathFilesDir,lengthCols,extra):
    listFiles = glob.glob(pathFilesDir+"*")
    #print(listFiles)
    #print(pathFilesDir+extra)
    fileMap = concatfile(listFiles,pathFilesDir+extra)
    frames = [] 
    frames.append(pd.read_csv(fileMap[0],header=0,usecols=list(range(1,lengthCols))))
    for i in range(1,len(listFiles)):
        frames.append(pd.read_csv(fileMap[i],header=0,usecols=list(range(1,lengthCols))))
    final = pd.concat(frames)
    final.reset_index(drop=True, inplace=True)
    return final    

    
```

```python
fullTweetsDF = mergefiles(tweetPath,11,"tweets")
fullTweetsDF.to_csv("../../data/1.stage2/tweets400k.csv")
fullTweetsDF.shape
fullTweetsDF.head()
```

```python
fullUsersDF = mergefiles(userPath,12,"users")
fullUsersDF.to_csv("../../data/1.stage2/users400k.csv")
fullUsersDF.shape
fullUsersDF.head()
```
